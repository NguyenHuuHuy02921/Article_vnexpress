```python
# Import neccessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics
```

```python
vectorized_data = pd.read_csv("combined_tfidf.csv")
vectorized_data
```

```python
# Read category from .csv file
category = pd.read_csv('category_labels.csv')
category['encoded_category'].value_counts()
```

```python
category.drop(['category'],axis=1, inplace=True)
```

```python
# Split data into train and test
X_train, X_test, Y_train, Y_test = train_test_split(vectorized_data, category, test_size=0.3, random_state=109)
X_train, X_test, Y_train, Y_test
```

```python
Y_test.value_counts()
```

```python
# Train the SVC model in different kernels
linear_kernel = SVC(kernel='linear')
linear_kernel.fit(X_train, Y_train)

rbf_kernel = SVC(kernel='rbf', gamma=0.5)
rbf_kernel.fit(X_train, Y_train)

sigmoid_kernel = SVC(kernel='sigmoid', gamma=0.7)
sigmoid_kernel.fit(X_train, Y_train)
```

```python
# Test model using test data
linear_predict = linear_kernel.predict(X_test)
rbf_predict = rbf_kernel.predict(X_test)
sigmoid_predict = sigmoid_kernel.predict(X_test)
```

```python
# Evaluate model with predicted result
print("Linear Model Result:")
print("Accuracy:", metrics.accuracy_score(Y_test, linear_predict))
print("Precision:")
print(metrics.precision_score(Y_test, linear_predict,average=None))
print("Recall:")
print(metrics.recall_score(Y_test, linear_predict,average=None))
print()

print("RBF Model Result:")
print("Accuracy:", metrics.accuracy_score(Y_test, rbf_predict))
print("Precision:", metrics.precision_score(Y_test, rbf_predict,average=None))
print("Recall:", metrics.recall_score(Y_test, rbf_predict,average=None))
print()

print("Sigmoid Model Result:")
print("Accuracy:", metrics.accuracy_score(Y_test, sigmoid_predict))
print("Precision:", metrics.precision_score(Y_test, sigmoid_predict,average=None))
print("Recall:")
metrics.recall_score(Y_test, sigmoid_predict,average=None)
```
